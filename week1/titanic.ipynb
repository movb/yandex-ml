```python
import pandas
data = pandas.read_csv('titanic.csv', index_col='PassengerId')
```

```python
data.head()
```

```python
def extract_name(full_name):
    name = full_name
    if name.find('(') != -1:
        name = name[name.find('(')+1:len(name)-1]
    elif name.find(',') != -1:
        name = name[name.find(',')+2:]
        if name[:4] == 'Mrs.':
            name = name[5:]
        elif name[:5] == 'Miss.':
            name = name[6:]
        elif name[:3] == 'Mr.':
            name = name[4:]
    return name.split()[0]
```

```python
name = 'Braund, Mr. Owen Harris'
extract_name(name)
```

```python
data['FirstName'] = data.Name.apply(extract_name)
```
